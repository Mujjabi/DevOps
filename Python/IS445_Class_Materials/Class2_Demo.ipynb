```python
df = pd.read_csv('building_inventory.csv', na_values={'Year Acquired': 0,
                                                      'Year Constructed': 0,
                                                      'Square Footage': 0
                                                      })
```

```python
df.describe()
```

```python
plt.plot(
    df['Year Acquired'],
    df['Year Constructed'])
```

```python
plt.plot(
    df['Year Acquired'],
    df['Year Constructed'],
    'o')
```

```python
plt.scatter(
    df['Year Acquired'],
    df['Year Constructed'])
```

```python
age = (df['Year Acquired'] - df['Year Constructed'])
```

```python
age
```

```python
df['Age at Acquisition'] = (df['Year Acquired'] - df['Year Constructed'])
```

```python
old_buildings = df['Age at Acquisition'] > 0
```

```python
old_df = df[old_buildings]
```

```python
old_df.loc[33:46]  # this is inclusive, not like slices
```

```python
old_df.iloc[0:1]  # this is like slices
```

```python
plt.hist("Age at Acquisition", data=old_df)
```

```python
plt.hist("Age at Acquisition", data=old_df, log=True)
```
