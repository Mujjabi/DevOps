```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
```

Basic exploration of the dataset

```python
# Displaying the first few rows of the dataset
print("First few rows of the dataset:")
df.head()
```

```python
# Displaying basic info about the dataset
print("\nDataset Info:")
print(df.info())
```

```python
# Displaying statistical summary for numerical columns
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```


The output indicates a summary of missing values in the Traffic Crashes dataset. It shows that several columns, such as 'CRASH_DATE_EST_I', 'LANE_CNT', 'INTERSECTION_RELATED_I', and others have a substantial number of missing entries. For example, 'CRASH_DATE_EST_I' and 'LANE_CNT' are missing in 725,634 and 585,541 cases, respectively. Conversely, many columns like 'CRASH_RECORD_ID', 'CRASH_DATE', and 'POSTED_SPEED_LIMIT' have complete data with zero missing values.

Exploring specific columns

```python
# Distribution of crashes by posted speed limit
plt.figure(figsize=(10,6))
sns.countplot(x='POSTED_SPEED_LIMIT', data=df, palette='viridis')
plt.title('Distribution of Crashes by Posted Speed Limit')
plt.xticks(rotation=90)  # This will rotate the labels to be vertical.
plt.tight_layout()  # This will adjust the plot to ensure everything fits without overlapping.
plt.show()
```

The bar chart illustrates the frequency of traffic crashes at various posted speed limits. A singular speed limit of 3 mph shows a markedly higher number of crashes compared to the others, indicating a notable concentration of incidents at that particular speed.

```python
# Weather conditions during crashes
plt.figure(figsize=(10,6))
sns.countplot(y='WEATHER_CONDITION', data=df, order = df['WEATHER_CONDITION'].value_counts().index)
plt.title('Weather Conditions during Crashes')
plt.show()
```


The image displays a horizontal bar chart titled "Weather Conditions during Crashes," which ranks various weather conditions by the number of associated traffic crashes. The 'Clear' condition has the highest count, indicating that most crashes occurred during clear weather. This is followed by significantly lower counts for 'Rain' and 'Unknown' conditions. Other weather conditions like 'Snow', 'Cloudy/Overcast', and several less common conditions such as 'Freezing Rain/Drizzle', 'Fog/Smoke/Haze', etc, are represented with progressively smaller counts. This visualization suggests that weather conditions are a factor in traffic crashes, with clear weather surprisingly showing the highest number of incidents.

```python
# Year VS crashes
# Converting 'CRASH_DATE' to datetime
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE'])

# Extracting the year from 'CRASH_DATE'
df['YEAR'] = df['CRASH_DATE'].dt.year

# Counting the number of crashes per year
yearly_crashes = df['YEAR'].value_counts().sort_index()

# Plotting the data
plt.figure(figsize=(10,6))
ax = sns.barplot(x=yearly_crashes.index, y=yearly_crashes.values, palette='Set3')
plt.title('Crashes by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45)  # Rotating the x-axis labels for better readability

plt.show()
```

The image depicts a bar chart titled "Crashes by Year," showing the number of crashes for each year from 2013 to 2023. Each year is represented by a colored bar, and the y-axis measures the 'Number of Crashes'.

There's a clear upward trend in the number of crashes from 2013 to 2018, with each year showing a higher count than the previous one. The year 2018 appears to have the highest number of crashes. After 2018, there is a slight decline or plateau in the number of crashes through to 2023.

The chart effectively shows the changes in the number of crashes over time, highlighting any trends or patterns in the data over the ten-year span.

Time Analysis

```python
# Crashes by hour of the day
plt.figure(figsize=(10,6))
sns.countplot(x='CRASH_HOUR', data=df, palette='coolwarm')
plt.title('Crashes by Hour of Day')
plt.show()
```

The image depicts a bar chart titled "Crashes by Hour of Day,". The bars represent the number of crashes that occurred at each hour throughout the day, with the hours ranging from 0 to 23, corresponding to a 24-hour clock.

This pattern suggests that traffic crashes are more frequent during the daytime and decrease as it gets later into the night. The peak hours likely correspond to busy traffic periods, such as morning rush hour, lunchtime, and evening rush hour, which are typically associated with increased vehicle activity and therefore a higher incidence of crashes.

```python
# Crashes by day of the week
plt.figure(figsize=(10,6))
ax = sns.countplot(x='CRASH_DAY_OF_WEEK', data=df, palette='Set2')
plt.title('Crashes by Day of Week')

# Defining the order of the days
days_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# Setting the x-axis labels to the days of the week
ax.set_xticklabels(days_of_week)

plt.show()
```

The image shows a bar chart titled "Crashes by Day of Week,". Each bar, colored differently for visual distinction, represents the total number of crashes that occurred on each day of the week, from Sunday to Saturday.

The distribution appears relatively even across the days, with no single day showing a drastically higher number of crashes than the others. However, there's a slight increase towards the end of the week, with Friday showing the highest count, followed by a small drop on Saturday. This pattern might suggest a correlation between the end-of-week activities, like commuting and increased leisure travel, and the frequency of crashes. The visual suggests that the risk of crashes is somewhat consistent, with slight variations on different days of the week.

```python
# Crashes by month
plt.figure(figsize=(10,6))
ax = sns.countplot(x='CRASH_MONTH', data=df, palette='Set3')
plt.title('Crashes by Month')

# Defining the order of the months
months_of_year = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Setting the x-axis labels to the month names
ax.set_xticklabels(months_of_year)
plt.xticks(rotation=90)  # This will rotate the labels to be vertical.
plt.tight_layout()

plt.show()
```


The image illustrates a bar chart titled "Crashes by Month,". Each bar, represented in a unique color for each month, shows the number of crashes recorded for each month of the year.

The distribution of crashes across months is relatively uniform, with no single month showing an overwhelmingly high number of crashes compared to the others. However, there seems to be a slight increase during the summer months, with the highest peaks in July and October. This could potentially be related to increased travel during these months or other seasonal factors affecting driving conditions. The chart provides a clear visual representation of the crash data across different times of the year.

Crash Analysis

```python
# Studying the severity of injuries in crashes
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='MOST_SEVERE_INJURY', order=df['MOST_SEVERE_INJURY'].value_counts().index)
plt.title('Severity of Injuries in Crashes')
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```

The image shows a bar chart titled "Severity of Injuries in Crashes,". The chart has five bars, each representing different categories of injury severity reported from traffic crashes.

The first bar, labeled 'NO INDICATION OF INJURY', is significantly taller than the others, indicating that the majority of crashes did not result in injuries. The subsequent bars represent increasing levels of injury severity: 'NONINCAPACITATING INJURY', 'REPORTED, NOT EVIDENT', 'INCAPACITATING INJURY', and 'FATAL'. Each of these categories has a progressively lower count, suggesting that serious injuries and fatalities occur less frequently than non-serious or non-injurious incidents.

This visual representation emphasizes that while most traffic crashes do not result in injury, a non-negligible number of incidents result in non-fatal injuries, and a smaller proportion result in fatalities.

```python
# Examining primary contributory causes of crashes
plt.figure(figsize=(12, 6))
cause_counts = df['PRIM_CONTRIBUTORY_CAUSE'].value_counts().head(10)  # Top 10 causes
sns.barplot(x=cause_counts.values, y=cause_counts.index, palette='rocket')
plt.title('Primary Contributory Causes of Crashes')
plt.xlabel('Number of Crashes')
plt.ylabel('Primary Contributory Cause')
plt.show()
```

The image shows a horizontal bar chart titled "Primary Contributory Causes of Crashes," which ranks various factors by their frequency as the primary cause of traffic crashes.

The longest bar corresponds to 'UNABLE TO DETERMINE', suggesting that for the majority of crashes, the primary cause could not be definitively identified. The other listed causes show a descending frequency, with 'FAILING TO YIELD RIGHT-OF-WAY' and 'FOLLOWING TOO CLOSELY' as the next most common contributing factors.

The chart visually communicates that a significant number of crashes are attributed to specific driving behaviors, but also highlights a substantial proportion where the cause is unclear. This data is critical for understanding and addressing the factors leading to traffic accidents.

```python
# Checking the correlation between injuries and crashes by month
injury_columns = ['INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT']
correlation = df[injury_columns + ['CRASH_MONTH']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='Blues')
plt.title('Correlation between Injuries and Crashes by Month')
plt.show()
```


The image depicts a heatmap titled "Correlation between Injuries and Crashes by Month," showing the correlation coefficients between different types of injuries and the month when crashes occurred. The types of injuries include total injuries, fatal, incapacitating, non-incapacitating, and reported but not evident injuries.

In the heatmap:

*  Darker shades of blue indicate a stronger positive correlation.
*  Lighter shades or white represent a weaker correlation.

The strongest correlations are seen within injury categories themselves, such as between total injuries and non-incapacitating injuries, indicated by the deep blue colors and higher correlation coefficients like 0.77. The correlation between injury categories and the crash month is quite weak, as shown by the very light shades of blue, with correlation coefficients close to zero.

The heatmap suggests that certain types of injuries are more likely to occur together in crashes, while the month of the crash does not strongly correlate with the frequency or severity of injuries.
